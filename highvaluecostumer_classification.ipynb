```python
#first we read the data
import pandas as pd
df = pd.read_csv('marketing_campaign.csv', sep='\t')
df.head()
```

```python
#we get some information of the data to know what we are dealing with and we get the number of nulls in the table
display(df.info())
display(df.describe())
display(df.isna().sum())
```

```python
# cleaning the data from null values as the previous table shown, by dropping the null rows although we could go any other way about this
df = df.dropna(subset=['Income'])
display(df.isna().sum())
```

```python
from datetime import datetime
#creating an age column for further analysis and comparisons
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']
display(df.head())
```

```python
#creating another column for further analysis and comparisons
df['Total_Spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
display(df.head())
```

```python
import matplotlib.pyplot as plt
#some plotting for visuals and easy analysis
plt.bar(df['Education'], df['Total_Spending'] )
plt.show()
```

we can notice that there is a huge gap between all education levels and basic level and this shows a clear area where focused marketing efforts could potentially yield significant results.

```python
import seaborn as sns
# Create age groups for the boxplot
df['Age_Group'] = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 70, 80, 90], labels=['20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89'])

# Create a boxplot of Total_Spending by Age Group
sns.boxplot(x='Age_Group', y='Total_Spending', data=df)
plt.title('Total Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Spending')
plt.show()
```

Median spending rises steadily from the 20-29 group up to the 80-89 group, so older customers tend to spend more on average.

```python
#the top 10 total spending per each age
df.groupby('Age')['Total_Spending'].sum().sort_values(ascending=False).head(10)
```

Ages 49 to 60 dominate the list with age 55 beign the highest total spending

```python
df['incom_group'] = pd.cut(df['Income'], bins=[0, 25000, 50000, 75000, 100000], labels=['Low', 'Medium', 'High', 'Very High'])
sns.boxplot(x='incom_group', y='Total_Spending', data=df)
plt.title('Total Spending by income group')
plt.xlabel('incom_group')
plt.ylabel('Total Spending')
plt.show()
```

this shows that the higher the income the more the total spending is, so focusing marketing efforts on high-income individuals is likely to yield higher returns in terms of total spending

```python
df['HighValueCustomer'] = df['Total_Spending'].apply(lambda x: 1 if x > df['Total_Spending'].median() else 0) #we can define what a valuable customer to us i did it this way
```

```python
from sklearn.model_selection import train_test_split
x = df[['Age','Education','Marital_Status','Kidhome','Teenhome','Income']] # we can use any related metric we want for a prediction
y = df['HighValueCustomer']
x = pd.get_dummies(x) # we did this because we have strings in the x data frame this replaces them with 0/1, for instance someone's education is phd in the education it will be 1 under phd
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#we import the random forrest classifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

The classification report and confusion matrix demonstrate that the model exhibits strong performance in identifying high-value customers, with an accuracy of 0.89. The low number of false positives and false negatives suggests the model is effective in identifying potential high-value customers thus efficient targeting, minimizing the misallocation of resources.
